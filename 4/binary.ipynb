```python
del binary_df['Lesson']
binary_df['Semester_Passed'] = binary_df['Semester_Passed'].map({'Even': 1, 'Odd': 0})
del binary_df['Instructor']
del binary_df['Total_ECTS_in_Semester']
del binary_df['Nth_Test_of_Semester']
binary_df['Homework'] = binary_df['Homework'].map({'Yes': 1, 'No': 0})
binary_df['Quiz'] = binary_df['Quiz'].map({'Yes': 1, 'No': 0})
binary_df['Passed'] = binary_df['Passed'].map({'Yes': 1, 'No': 0})
binary_df['Conditionally_Passed'] = binary_df['Conditionally_Passed'].map({'Yes': 1, 'No': 0})
del binary_df['num_Grade_values']
```

```python
min_sup = 0.50
min_conf = 0.70

df_encoded = pd.get_dummies(binary_df)

frequent_itemsets = apriori(df_encoded, min_support=min_sup, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_conf)
```
